```python
import pickle
import pandas as pd
```

```python
participant_data = pickle.load(open('../projects.p', 'rb'))
```

```python
participant_data
```

```python
participants = []

for project in participant_data.keys():
    for login in participant_data[project].keys():
        data = participant_data[project][login]
        participants.append({
            'project': project,
            'login': login,
            'commits': data['commits'],
            'reports': data['issues']
        })

df = pd.DataFrame(participants)
df
```

```python
df.to_pickle('participants.p')
```
